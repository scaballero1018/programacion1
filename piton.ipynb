```python
# Series de Fibonacci:
# la suma de dos elementos define el siguiente
a, b = 0, 1
while b < 100:
  print(b)
  a, b = b, a+b #el valor es de b, de b no esta actualizado
```

```python
a, b = 0, 1
for b in range(20):   # 6 repeticiones
    print(b)
    a, b = b, a + b
```
