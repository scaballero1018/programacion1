```python
# Series de Fibonacci:
# la suma de dos elementos define el siguiente
a, b = 0, 1
while b < 100:
  print(b)
  a, b = b, a+b #el valor es de b, de b no esta actualizado
```

```python
for n in range(0, 10):
  if n == 1 or n == 0:
    print("El número es ", n)
  else:
    for x in range(2, n):
      if n % x == 0:
          print(n, 'es igual a', x, '*', n // x)
          break
    else: # Este else corresponde al for, no al if
      print(n, 'es un número primo')

```

```python
lista = (10, 20, 30, 40, 50)
try:
    index = int(input("Ingresa un índice: "))
    print("El valor es:", lista[index])
except IndexError: # Indice fuera de rango
    print("Índice fuera de rango")
except ValueError: # Cuando ingresamos un valor no permitido
    print("Debes ingresar un número entero")
```
