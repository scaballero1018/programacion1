```python
def analizar_texto(texto):
  mayusv=minusv=mayusc=minusc=numeros=espacios=otros=0
  vocales="aeiouAEIOU"
  consonantes="bcdfghjklmnñpqrstvwxyzBCDFGHJKLMNÑPQRSTVWXYZ"
  for i in texto:
    if i in vocales:
      if i.isupper():
        mayusv+=1
      else:
        minusv+=1
    elif i in consonantes:
      if i.isupper():
            mayusc+=1
      else:
            minusc+=1
    elif i.isdigit():
      numeros+=1
    elif i.isspace():
      espacios+=1
    else:
      otros+=1
  return {"mayúsculas vocales": mayusv,
          ", minúsculas vocales": minusv,
          "mayúsculas consonantes": mayusc,
          "minúsculas consonantes": minusc,
          "números": numeros,
          "espacios": espacios,
          "otros":otros}


```

```python
print(analizar_texto("HOla"))
```
